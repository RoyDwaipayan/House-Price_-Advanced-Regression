```python
data_train_target = data_train['SalePrice']
data_4 = pd.get_dummies(data_train, drop_first = True)
data_4 = data_4.drop(['SalePrice','Utilities_NoSeWa', 'Condition2_RRAe', 
             'Condition2_RRAn', 'Condition2_RRNn', 'HouseStyle_2.5Fin', 
             'RoofMatl_CompShg', 'RoofMatl_Membran', 'RoofMatl_Metal', 'RoofMatl_Roll', 
             'Exterior1st_ImStucc', 'Exterior1st_Stone', 'Exterior2nd_Other', 'Heating_GasA', 'Heating_OthW', 
             'Electrical_Mix', 'GarageQual_Fa', 'PoolQC_Fa', 'MiscFeature_TenC'],axis=1)
print('Train data shape: ',data_4.shape)

data_4_test = pd.get_dummies(data_test, drop_first = True)
data_4_test = data_4_test[data_4.columns.values]
print('Test data shape: ',data_4_test.shape)
```

```python
# fill missing values based on probability of occurrence
for column in data_4.columns:
    null_vals = data_4.isnull().values
    a, b = np.unique(data_4.values[~null_vals], return_counts = 1)
    data_4.loc[data_4[column].isna(), column] = np.random.choice(a, data_4[column].isnull().sum(), p = b / b.sum())
    
# fill missing values based on probability of occurrence
for column in data_4_test.columns:
    null_vals = data_4_test.isnull().values
    a, b = np.unique(data_4_test.values[~null_vals], return_counts = 1)
    data_4_test.loc[data_4_test[column].isna(), column] = np.random.choice(a, data_4_test[column].isnull().sum(), p = b / b.sum())
```
